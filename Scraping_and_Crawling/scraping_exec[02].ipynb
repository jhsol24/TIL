```python
from urllib.request import urlopen
from bs4 import  BeautifulSoup
from urllib.request import urlopen
from urllib.error   import HTTPError
from urllib.error   import URLError
```

```python
try:
    html = urlopen('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx')
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser')
```

```python
soup
```

```python
table = soup.find('table', {'class':'tData01 tt'})
table
```

```python
data = []
name_list = []
avg_list = []
hit_list = []
hr_list = []
rbi_list = []

for tr in table.find_all('tr') :
    tds = list(tr.find_all('td'))
    
    for td in tds :
        if td.find('a'):
            
            name = td.find('a').text
            name_list.append(name)
            avg = tds[3].text
            avg_list.append(avg)
            hit = tds[8].text
            hit_list.append(hit)
            hr = tds[11].text
            hr_list.append(hr)
            rbi = tds[13].text
            rbi_list.append(rbi)
            data.append([name, avg, hit, hr, rbi])
```

```python
with open('baseballplayer.csv', 'w',  encoding='UTF-8') as file:
    # 컬럼 이름 추가
    file.write('player, AVG, Hit, HomeRun, RBI\n')
    
    # 데이터를 반복하면서 선수이름, 타율, 안타수를 줄 단위로 저장
    for idx in data :
        file.write('{},{},{},{},{}\n'.format(idx[0], idx[1], idx[2], idx[3], idx[4]))
print('ok')
```

```python
baseballPlayer_df = pd.DataFrame({
    'name' : name_list,
    'avg' : avg_list,
    'hit' : hit_list,
    'hr' : hr_list,
    'rbi' : rbi_list
})
```
