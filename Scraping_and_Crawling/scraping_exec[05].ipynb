# 왜 크롤링이 필요한지 크롤링 준비 & 이미지 스크래핑 & 자연어 처리
>코로나 이미지 검색하여 50개를 이미지 파일로 저장하기

```python
from urllib.request import urlopen
from bs4 import  BeautifulSoup
from urllib.request import urlopen
from urllib.error   import HTTPError
from urllib.error   import URLError
from urllib.parse   import quote_plus
```

```python
base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EC%BD%94%EB%A1%9C%EB%82%98'
keyword = input('검색어 입력 : ')
imgCnt   = int(input('스크랩 할 이미지 개수 : '))

try:
    url  = base_url+quote_plus(keyword)
    html = urlopen(url)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser', from_encoding='utf-8')
```

```python
img = soup.find_all('img', '_img')
img
```

```python
cnt = 1
for i in img :
    imgUrl = i['data-source']
#     print(imgUrl)   # 이미지의 link만 골라서 출력함
    with urlopen(imgUrl) as file :
        with open('./images/'+str(cnt)+'.jpg', 'wb') as imgFile :
            img = file.read()
            imgFile.write(img)
            
    cnt += 1
    if cnt > imgCnt :
        break
        
print('image download success')

# 이미지를 다운로드 해서 저장한다.
```

## csv 파일로 pandas의 DataFrame을 활용해서 저장하기

```python
base_url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query='
keyword = input('검색어 입력 : ')

try:
    url  = base_url+quote_plus(keyword)
    html = urlopen(url)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser', from_encoding='utf-8')
```

```python
# sh_blog_title
titles=soup.find_all('a',{'class':'sh_blog_title'})
titles
```

##  제목, 링크 가져와서

## csv 파일 만들어 보는 실습 (DataFrame)

## csv 파일 읽기까지

```python
title_list = []
link_list  = []
for i in titles :
    print(i.attrs['title'])   # attrs : 내부에 있는 항목들을 리스트로 가져온다.
    print(i.attrs['href'])
    print()
    title_list.append(i.attrs['title'])
    link_list.append(i.attrs['href'])
```

```python
print(len(title_list))
print(len(link_list))
```

```python
import pandas as pd
python_df = pd.DataFrame({
    'title' : title_list,
    'link'  : link_list
})
```

```python
python_df.to_csv('./data/python_df.csv', mode='w', encoding='utf-8', index=False)
print('success')
```

## 크롤링을 통한 페이지 처리 방법 익히기

### 크롤러

```python
from time import sleep
from random import randint
import pandas as pd
from IPython.core.display import clear_output
from time import time
from bs4 import BeautifulSoup
from requests import get
```

```python
# loop 사이에 잘 진행되고 있는지를 보기위한  indicator sentence
start_time = time()
# start_time
requests = 0
for _ in range(5) :
    requests += 1
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
    print("request : {}, frequency : {}".format(requests, requests/elapsed_time))
    clear_output(wait=True)
```

```python
base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EC%BD%94%EB%A1%9C%EB%82%98'
keyword = input('검색어 입력 : ')
imgCnt   = int(input('스크랩 할 이미지 개수 : '))

try:
    url  = base_url+quote_plus(keyword)
    html = urlopen(url)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser', from_encoding='utf-8')
```

```python
# sh_blog_title
titles=soup.find_all('a',{'class':'sh_blog_title'})
titles
```

```python
pages = [str(i) for i in range(1,11)]
pages
```

```python
price_list = []
title_list = []
link_list  = []


# loop 사이에 잘 진행되고 있는지를 보기위한  indicator sentence
start_time = time()
requests = 0
for page in pages :
    requests += 1
    url = get('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%EA%B7%B8%EB%9E%98%ED%94%BD%EC%B9%B4%EB%93%9C&pagingIndex=' + page + '&pagingSize=40&productSet=total&query=%EA%B7%B8%EB%9E%98%ED%94%BD%EC%B9%B4%EB%93%9C&sort=rel&timestamp=&viewType=list')
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
#     print("request : {}, frequency : {}".format(requests, requests/elapsed_time))
#     clear_output(wait=True)
    soup = BeautifulSoup(url.text, 'html.parser')
#   title, link, price
#   가격 가져오기
    prices = soup.select('.price_num__2WUXn')
    prices_text = [price.text for price in prices]
    price_list.extend(prices_text)
#   타이틀, 링크 가져오기
    titles = soup.select('.basicList_link__1MaTN')
    titles_text = [title.text for title in titles]
    title_list.extend(titles_text)

    link_text = [title['href'] for title in titles]
    link_list.extend(link_text)
```

```python
print(len(price_list))
print(len(title_list))
print(len(link_list))
```

```python
pythonShopping_df = pd.DataFrame({
    'title' : title_list,
    'price' : price_list,
    'link' : link_list
})
pythonShopping_df
```

```python
pythonShopping_df.to_csv('./data/pythonShopping_df.csv', mode='w', encoding='utf-8', index=False )
print('success')
```
