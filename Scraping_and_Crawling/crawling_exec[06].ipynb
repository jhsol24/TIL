## Json 형식의 파일 크롤링

```python
import json, re
from urllib.request import urlopen
from html import unescape
```

```python
# 웹 페이지 읽어오기
request = urlopen('http://www.hanbit.co.kr/store/books/full_book_list.html')
encoding = request.info().get_content_charset('utf-8')
html = request.read().decode(encoding)
html
```

>json.dumps() : 데이터를 json 형태로 인코딩 (문자열)
    
>json.dump() : 데이터를 json 형태로 인코딩하여 파일에 출력 (파일)

>ensure_ascii = False : \xxxxx 형태로 이스케이프 하지 않고 출력

>[{key : value}, {key : value}]indent=size

```python
### 정규표현식
- . 모든 문자

- * 0번 이상 반복

- ? 있어도 되고 없어도 된다.

- <a href="(.*?)">
    
- <td class='left'><a.*?</td>
```

## json 파일 생성
> 목표 : json 형식을 익히고 정규표현식 사용 익히기

```python
with open('./data/booklist_json.json', mode='w', encoding='utf-8') as file :
    data = []
    for partial_html in re.findall('<td class="left"><a.*?</td>', html) :
#         print(partial_html)
        search = re.search(r'<a href="(.*?)">', partial_html).group(1)
#         안에 공백이면 a 태그까지, 1 넣으면 내용만
#         print(search)
        url = 'https://www.hanbit.co.kr'+search
#         print(url)
        title = re.sub(r'<.*?>', '', partial_html)
#         print(title)
        data.append({'bookName' : title, 'link' : url})
        print(json.dumps(data, ensure_ascii=False, indent=2))
#         url과 title을 json 형식으로 가져오는 것을 중점으로 함
    json.dump(data, file, ensure_ascii=False, indent=2)
```

# 바이크 제조사 페이지 크롤링 [실습]

>바이크 제조사 정보를 다음 페이지 버튼을 활용해 크롤링한다.

>- 전체 제조사 엘리먼트를 찾아서 마우스 클릭하고 그 결과물을 반환받는다
- 바이크 제조사 정보를 크롤링
- 다음 페이지 버튼이 활성화 되어 있으면 클릭하고 추가적으로 제조사 정보를 반환받는다
- 다음 페이지 버튼이 비활성화 되어 있으면 크롤링을 중단한다.

## 바이크 제조사 전체 페이지 클릭 구현

```python
bikeElementBtn = driver.find_element_by_css_selector('#container > div.spot_main > div.spot_aside > div.tit > a')
# 태그에 대한 경로 설정은 다양하게 할 수 있다. 겹치지 않을 정도로만 해주면 된다.
bikeElementBtn.click()
```

```python
time.sleep(3)   # 계속 빨리 돌면 웹페이지에 대한 공격으로 인지하여 시간을 주고 돌린다.
```

### 바이크 제조사 1페이지 바이크 리스트 추출

```python
bikeElements = driver.find_elements_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.emblem_area > ul > li')
print(bikeElements)
```

```python
for bike_li in bikeElements :
    companyName = bike_li.find_element_by_tag_name('span').text
    
    if (companyName != '') :
        print(companyName.strip())
        # image
        img = bike_li.find_element_by_tag_name('img').get_attribute('src')
        print(img)
        # link
        link = bike_li.find_element_by_tag_name('a').get_attribute('href')
        print(link)
        
time.sleep(3)
```

### 제조사 리스트 버튼을 클릭하는 부분 찾기

```python
nextBtn = driver.find_element_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.rolling_btn > button.next')
isFlag   = nextBtn.is_enabled()
print(isFlag)
```

```python
if(isFlag == True) :
    print('next page exist~~~')
    nextBtn.click()
    bikeElements = driver.find_elements_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.emblem_area > ul > li')
    for bike_li in bikeElements :
        companyName = bike_li.find_element_by_tag_name('span').text
    
        if (companyName != '') :
            print(companyName.strip())
            # image
            img = bike_li.find_element_by_tag_name('img').get_attribute('src')
            # get_attribute(*) : 요소 속성중 *에 해당하는 속성 값을 추출
            
            print(img)
            # link
            link = bike_li.find_element_by_tag_name('a').get_attribute('href')
            print(link)
```
