## Pie chart 
>카테고리별 값의 상대적 비교를 해야할 경우
- 차트 원형을 유지 : plt.axis('equal')

```python
labels = ['뽕사부','순대국밥','탐라도야지','바른식탁','서브웨이']
sizes =  [15, 15, 20, 25, 25]
colors = ['gold','purple','blue','yellow','gray']

plt.title('Pie Chart')
plt.pie(sizes, labels = labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

## Histogram 
>구간에 대한 데이터 집계를 시각화 할 때 사용

```python
data = np.random.randn(1000)
data

plt.title('histogram')
plt.hist(data, bins=20)
plt.grid()
plt.show()
```

## Scatter plot
>데이터 집합의 상관관계를 살펴보기위한 차트
- 2차원 데이터 표시

```python
np.random.seed(100)
X = np.random.normal(0,1,100)
Y = np.random.normal(0,1,100)
plt.title(scatter plot')
plt.scatter(X,Y)
plt.show()
```

```python
np.random.seed(100)
X = np.random.rand(30)
Y1 = np.random.rand(30)
Y2 = np.random.rand(30)
Y3 = np.pi * (15 * np.random.rand(30))**2

plt.title('bubble chart')
plt.scatter(X, Y1, c=Y2, s=Y3)
plt.show()
```

### autopct 사용법
- '%0.2f' — 소숫점 두번째자리 까지 보여준다. 4.08
- '%0.2f%%' — 소숫점 두번째자리까지 보여주고 맨 뒤에 ‘%’ 기호를 붙인다. 4.08%.
- (% 기호를 두 번 쓰는 이유는 문자열 포매팅 할 때 처음 % 기호가 이스케이프 문자처럼 작동하기 때문이다.)
- '%d%%' — 가장 가까운 정수로 반올림하고 맨 뒤에 % 기호를 붙인다.


```python
titanic.head()
# 선실별 승객 수 비율을 pie chart로 구현
titanic_pclass = pd.DataFrame(titanic['pclass'].value_counts())
titanic_pclass

labels = [3, 1, 2]
colors = ['gold','purple','blue']

plt.title('The number of passengers by pclass')
plt.pie(titanic_pclass, labels = labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()


# 강사님 답안
pclass_ratio_cnt = titanic.pclass.value_counts()
# type(pclass_ratio_cnt)
pclass_ratio_cnt.plot.pie(autopct='%.2f%%',
                          startangle = 45,
                          legend=True,
                          shadow=True)
plt.axis('equal')
plt.show()
```
