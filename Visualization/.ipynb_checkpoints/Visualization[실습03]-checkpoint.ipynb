```python
files = glob('./data/지역*.xls')
files
```

```python
tmp_raw = []
for file in files :
    tmp = pd.read_excel(file, header=2)
    tmp_raw.append(tmp)
# tmp_raw
```

```python
oil_station = pd.concat(tmp_raw)
oil_station.head()
```

```python
oil_station.info()
```

```python
# 휘발유만 대상으로 분석 진행
stations = pd.DataFrame({'oil_store' : oil_station['상호'],
                         'address' : oil_station['주소'],
                         'price' : oil_station['휘발유'],
                         'self_y_n' : oil_station['셀프여부'],
                         'company' : oil_station['상표']})
stations
```

```python
# 파생변수 생성
# 구별 주유 가격을 조사하기 위해
stations['direct'] = [addr.split()[1] for addr in stations['address']]
stations.head()
```

```python
# 구별 정보 확인
stations['direct'].unique()
```

```python
# 서울특별시 -> 성동구
# 특별시 -> 도봉구 변경하기
# stations[stations['direct']=='서울특별시']
# stations[stations['direct']=='서울특별시']

stations.loc[stations['direct']=='서울특별시', 'direct'] = '성동구'
```

```python
stations.loc[stations['direct']=='특별시', 'direct'] = '도봉구'
```

```python
# 구별 정보를 확인
stations['direct'].unique()
```

```python
stations[ stations['price'] == '-' ]
```

```python
# 가격정보가 기입되지 않은 주유소는 제거

stations = stations[stations['price'] != '-']
```

```python
# 컬럼 타입을 숫자형 변경
stations['price'] = [float(value) for value in stations['price']]
```

```python
stations.reset_index(inplace=True)
# stations.drop('index', axis=1)
del stations['index']
```

```python
stations.info()
```

```python
stations.boxplot(column='price', by='self_y_n', figsize=(12,8))
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='company', y='price', hue='self_y_n', data=stations, palette='Set3')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='company', y='price', data=stations, palette='Set3')
sns.swarmplot(x='company', y='price', data=stations, color='.6')
plt.show()
```

```python
# 피봇 이용해서 구별 가격 평균값 생성
direct_pivot = pd.pivot_table(stations, index='direct', values='price', aggfunc=np.mean)
direct_pivot.head()
```

```python
import json
import folium
import googlemaps
# 구별 정보에 대해서 googlemaps 활용해서 시각화
geo_str = json.load( open('./data/02. skorea_municipalities_geo_simple.json', encoding='utf-8'))
```

```python
!pip show folium
```

```python
map = g.Map(location=[37.55, 126.98], zoom_start=10.5, tiles='Stamen Terrain')
map.choropleth(geo_str,
              data = direct_pivot,
              columns=[direct_pivot.index, 'price'],
              fill_color='PuRd',
              key_on='feature.id')

map
```

```python
# 주유가격 상위 10개, 하위 10개 주유소를 맵으로 시각화
oil_station_top10 = stations.sort_values(by='price', ascending=False)[:10]
display(oil_station_top10)

oil_station_bottom10 = stations.sort_values(by='price')[:10]
display(oil_station_bottom10)
```

```python
gmap_key = 'A'
gmaps = googlemaps.Client(key=gmap_key)
```

```python
lat = []
lng = []
for n in oil_station_top10.index:
    tmp_addr = str(oil_station_top10['address'][n])
    # print(tmp_addr)
    tmp_map = gmaps.geocode(tmp_addr)
    # print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')

    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])
    
print(lat)
print(lng)

oil_station_top10['lat'] = lat
oil_station_top10['lng'] = lng

oil_station_top10
```

```python
for name, lat, lng in zip(oil_station_bottom10.oil_store, oil_station_bottom10.lat, oil_station_bottom10.lng):
    marker = folium.Marker([lat, lng], popup=name, icon=folium.Icon(color='blue'))
    marker.add_to(map)
    
map
```
