```python
seoul_school_df = pd.read_excel("./data/elem_school_folium.xlsx")
seoul_school_df.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명, 위도, 경도
feature_sel_df = seoul_school_df[['학교명','학교급구분','소재지도로명주소','위도','경도']]
feature_sel_df.head()
```

```python
# 소재지도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출

# 결측값 확인, 제거
print(feature_sel_df.isnull().sum())
feature_sel_df.dropna(inplace=True)
print('*'*77)
print(feature_sel_df.isnull().sum())
```

```python
feature_sel_df[feature_sel_df['소재지도로명주소'].str.contains('서울')].head()
```

```python
# 방법 1
# 조건에 부합하는 결과를 데이터 프레임으로 만들기 위해서 query()

# query() : Pandas에서 DataFrame에서 조건에 맞는 데이터를 가져오는 것.
# loc나 []과 비교해 문자열로 입력 가능하고 빠르다는 장점이 있다.

condition = "학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울특별시').values"
result_df = feature_sel_df.query(condition)
result_df.head()
```

```python
# 방법 2
# 인덱스 추출
index_seoul = feature_sel_df['소재지도로명주소'].str.contains('서울특별시')
# type(index_seoul)
index_ele = feature_sel_df['학교급구분'] == '초등학교'
print(len(index_seoul))
print(len(index_ele))

result_df = feature_sel_df.loc[index_seoul & index_ele]
result_df.tail()
```

```python
school_map = g.Map(location=[37.55, 126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

# 지도 시각화
for name, lat, lng in zip(result_df.index, result_df.위도, result_df.경도):
    g.CircleMarker([lat, lng],
             popup=name, 
             icon=g.Icon(color='skyblue'),
             radius=10,
             fill=True,
             fill_color='coral',
             fill_opacity=0.3).add_to(school_map)
    
school_map
```

# json 파일 활용 경기도 경계 정보 가져오기

```python
# 경계정보를 가진 json 파일 읽어보자
geo_data = json.load(open('./data/경기도행정구역경계.json', encoding='utf-8'))
# json 열 때는 load
# geo_data
```

```python
pop_df.columns = pop_df.columns.map(str)
# 지도생성
gyeonggi_map = g.Map(location=[37.503539, 127.025336],
                    zoom_start=8,
                    tiles='Stamen Terrain')

# 구분된 경계를 표시할 때 사용하는 객체
year = '2017'
g.Choropleth(geo_data = geo_data,
            data = pop_df[year],
            columns = [pop_df.index, pop_df[year]],
            key_on = 'feature.properties.name').add_to(gyeonggi_map)
gyeonggi_map
```
