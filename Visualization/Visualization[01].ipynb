```python
# 문제 02
# 자동차 제조회사에 따라 도시 연비가 다른지 알아보려고 한다.
# audi와 toyota 중 어느 manufacturer(자동차 제조 회사)의 cty(도시연비)가
# 평균적으로 더 높은지 알아보기

# solution 1
print( data_df[data_df['manufacturer'] == 'audi']['cty'].mean() )
print( data_df[data_df['manufacturer'] == 'toyota']['cty'].mean() )

# solution 2
print( data_df.query('manufacturer == "audi"')['cty'].mean() )
print( data_df.query('manufacturer == "toyota"')['cty'].mean() )
```

```python
# 문제 03
# chevrolet, ford, honda 세 자동차의 고속도로 연비 평균을 알아보려고 한다.(각각 X)
# 이 회사들의 데이터를 추출한 후 hwy 전체 평균을 확인하기

# solution 1
manufac_car = ['chevrolet','ford','honda']
print( data_df.query('manufacturer in @manufac_car')['hwy'].mean() )

# solution 2
print( data_df.query('manufacturer == "chevrolet" or \
                      manufacturer == "ford" or \
                      manufacturer == "honda"')['hwy'].mean() )

# solution 3
print( data_df[ (data_df['manufacturer'] == 'chevrolet') |
                (data_df['manufacturer'] == 'ford') |
                (data_df['manufacturer'] == 'honda') ]['hwy'].mean() )
```

```python
chevrolet_index = data_df['manufacturer'] == 'chevrolet'
ford_index      = data_df['manufacturer'] == 'ford'
honda_index     = data_df['manufacturer'] == 'honda'

temp_df = data_df.loc[chevrolet_index | ford_index | honda_index, :]
mean_df = temp_df[ ['manufacturer', 'hwy'] ].groupby(temp_df.manufacturer).agg(np.mean)
mean_df.plot.bar()

display(mean_df)
print('세 자동차 회사의 hwy의 평균 : ', temp_df['hwy'].mean())
```

### 데이터 전처리 : 연료 가격 ***변수 추가***
- c : 가스 2.35, d : 디젤 2.38, e : 에탄올 2.11, p : 고급 휘발유 2.76, r : 보통 휘발유 2.22

```python
fl_df = pd.DataFrame({
    'fl' : ['c','d','e','p','r'],
    'price_fl' : [2.35, 2.38, 2.11, 2.76, 2.22]
})
fl_df
```

```python
# 연료 가격 프레임을 원본 프레임에 병합 (merge)
data_df01 = pd.merge(data_df, fl_df, on='fl', how='inner')
data_df01
```

```python
# 병합한 데이터프레임에서  model,fl,price_fl 만 추출
data_df01[['model','fl','price_fl']].head()
```

```python
data_df01.isna().sum()
```

```python
# 구동방식(drv) 별 고속도로 연비(hwy) 평균
# 임의적으로 결측값 처리를 위해서 더미 값을 넣어보기
data_df01.loc[65, 'hwy'] = np.nan
data_df01.loc[120, 'hwy'] = np.nan
data_df01.loc[154, 'hwy'] = np.nan
data_df01.loc[189, 'hwy'] = np.nan
data_df01.loc[219, 'hwy'] = np.nan
data_df01.loc[230, 'hwy'] = np.nan
```

```python
# 데이터 전처리 과정에서 결측값 확인하기
data_df01.isnull().sum()
```

```python
# hwy 변수의 결측값을 제외하고, 어떤 구동방식의 고속도로 평균 연비가 높은지 알아보기
data_df02 = data_df01.filter(['drv','hwy']).dropna()
print(data_df02.info())

drv_group_df = data_df02.groupby('drv').mean()
display(drv_group_df)
```

```python
# 구동방식별 연비평균을 비교하는 막대 그래프 시각화
drv_group_df.plot.bar(rot=0)
plt.title('구동방식별 평균 연비')
plt.xlabel('구동방식')
plt.ylabel('연비')
plt.grid()
plt.show()
```

```python
# 20201020
# 이상치 정제

# car_mpg 데이터에서 구동방식별 고속도로 연비 평균과 도시연비 평균을 극단치를 제외하고 확인
# 각 연비별 이상치 확인 : boxplot()

outlier_df = data_df.filter(['hwy','cty'])
outlier_df.boxplot()

plt.show() # 안해줘도 결과 나오지만 해주는 습관을 가지자.
```

```python
# 3사분위 정보 추출
quantile75 = outlier_df.quantile(q=0.75)
quantile75
```

```python
# 1사분위 정보 추출
quantile25 = outlier_df.quantile(q=0.25)
quantile25
```

```python
# IQR (3사분위 수 - 1사분위 수)
iqr = quantile75 - quantile25
iqr
```

```python
# 최저 한계치 (lower fence)
lower_fence = quantile25 - 1.5 * iqr
print('lower_fence\n', lower_fence)

# 최고 한계치 (upper fence)
upper_fence = quantile75 + 1.5 * iqr
print('upper_fence\n', upper_fence)
```

```python
# outlier

# 극단치 경계 값 (아래 수평선)
lower_outlier = outlier_df[outlier_df > lower_fence].min()
print(lower_outlier)
```

```python
# 극단치 경계 값 (위 수평선)
upper_outlier = outlier_df[outlier_df < upper_fence].max()
print(upper_outlier)
```

```python
# 이상치를 결측값으로 변경
# 고속도로 연비의 이상치 데이터 추출
hwy_outlier_df = data_df.query('hwy > '+str(upper_outlier['hwy']))
hwy_outlier_df
```

```python
# 도시 연비의 이상치 데이터 추출
cty_outlier_df = data_df.query('cty > '+str(upper_outlier['cty']))
cty_outlier_df
```

```python
# copy()
outlier_clean_df = data_df.copy()
outlier_clean_df.head()
```

```python
# hwy 연비에 대한 이상치를 결측값으로 변경
for idx in hwy_outlier_df.index :
#     print(idx)
    outlier_clean_df.loc[idx, 'hwy'] = np.nan
outlier_clean_df.head()
```

```python
# cty 연비에 대한 이상치를 결측값으로 변경
for idx in cty_outlier_df.index :
#     print(idx)
    outlier_clean_df.loc[idx, 'cty'] = np.nan
outlier_clean_df.head()
```

```python
# 결측값 있는지 확인
outlier_clean_df.filter(['cty','hwy']).isna().sum()
```

```python
# 결측값 제거한 구동방식별 고속도로, 도시 연비 평균
drv_mean_df = outlier_clean_df.filter(['drv','cty','hwy']).dropna().groupby('drv').mean()
drv_mean_df
```

```python
drv_mean_df.plot.bar(rot=0)
plt.grid()
plt.title('이상치 제거 후 고속도로, 도시 평균 연비')
plt.xlabel('구동방식')
plt.ylabel('평균 연비')
plt.show
```
